# 1. ¿Por qué Pandas?

Pandas permite trabajar con datos tabulares de forma más expresiva que listas o diccionarios puros.  
En Data Science, normalmente el flujo inicia en una tabla, pasa por limpieza y feature engineering, y termina en reportes, modelos o dashboards.

## 1.1 De Python puro a Pandas

Pregunta aplicada:  
**¿Cuánto revenue generó cada producto?**

```python
ventas = [
    {"producto": "Prestamo Personal", "cantidad": 10, "precio": 120.0},
    {"producto": "Tarjeta Credito", "cantidad": 4, "precio": 80.0},
    {"producto": "Prestamo Personal", "cantidad": 7, "precio": 120.0},
    {"producto": "Seguro", "cantidad": 3, "precio": 45.0},
]

# Python puro
total_por_producto = {}
for v in ventas:
    producto = v["producto"]
    total_por_producto[producto] = total_por_producto.get(producto, 0) + v["cantidad"] * v["precio"]

total_por_producto
```

```python
# Pandas
df_demo = pd.DataFrame(ventas)

resumen_demo = (
    df_demo
    .assign(revenue=lambda d: d["cantidad"] * d["precio"])
    .groupby("producto", as_index=False)
    .agg(revenue=("revenue", "sum"))
    .sort_values("revenue", ascending=False)
)

resumen_demo
```

# 2. Caso de clase: campañas comerciales

Trabajaremos con dos tablas:

- `clientes`: información maestra del cliente.
- `campanas_raw`: comunicaciones comerciales por cliente, canal y producto.

Este diseño refleja un flujo típico de Data Science: una tabla transaccional se enriquece con una tabla maestra para construir métricas, features y targets.

## 2.1 Construcción de datos sintéticos realistas

Incluiremos problemas comunes:

- Fechas como texto.
- Montos con símbolos y formatos mixtos.
- Categorías sucias: espacios, mayúsculas, abreviaturas.
- Valores faltantes.
- Duplicados.
- Outliers.
- Variables listas para análisis comercial y modelamiento.

```python
n_clientes = 600
n_contactos = 2_500

clientes = pd.DataFrame({
    "cliente_id": np.arange(100_000, 100_000 + n_clientes),
    "edad": rng.integers(18, 72, size=n_clientes),
    "sexo": rng.choice(["F", "M"], size=n_clientes, p=[0.52, 0.48]),
    "region": rng.choice(["Lima", "Norte", "Sur", "Centro", "Oriente"], size=n_clientes, p=[0.46, 0.18, 0.14, 0.12, 0.10]),
    "segmento_cliente": rng.choice(["Masivo", "Emergente", "Premium"], size=n_clientes, p=[0.62, 0.28, 0.10]),
    "antiguedad_meses": rng.integers(1, 96, size=n_clientes),
    "ingreso_mensual": rng.lognormal(mean=8.0, sigma=0.45, size=n_clientes).round(2),
    "score_digital": rng.beta(2.5, 2.0, size=n_clientes).round(4),
})

# Algunos ingresos faltantes y extremos para simular datos reales
clientes.loc[rng.choice(clientes.index, size=18, replace=False), "ingreso_mensual"] = np.nan
clientes.loc[rng.choice(clientes.index, size=5, replace=False), "ingreso_mensual"] *= 7

clientes.head()
```

```python
fechas = pd.date_range("2025-01-01", "2025-06-30", freq="D")

campanas_raw = pd.DataFrame({
    "contacto_id": np.arange(1, n_contactos + 1),
    "fecha_contacto": rng.choice(fechas, size=n_contactos, replace=True).astype(str),
    "cliente_id": rng.choice(clientes["cliente_id"], size=n_contactos, replace=True),
    "canal": rng.choice(["TLV", "tlv ", "HTML", "html ", "SMS", "sms", "APP", "app "], size=n_contactos, p=[0.18, 0.04, 0.23, 0.05, 0.24, 0.06, 0.16, 0.04]),
    "producto": rng.choice(["PP", "TC", "Extracash", "Seguro"], size=n_contactos, p=[0.42, 0.30, 0.18, 0.10]),
    "monto_ofertado": rng.lognormal(mean=9.1, sigma=0.55, size=n_contactos).round(2),
    "score_modelo": rng.beta(2.2, 5.5, size=n_contactos).round(4),
})

# Enriquecemos temporalmente con atributos para simular una respuesta comercial coherente
_tmp = campanas_raw.merge(clientes[["cliente_id", "score_digital", "segmento_cliente"]], on="cliente_id", how="left")

canal_effect = _tmp["canal"].str.strip().str.upper().map({"TLV": 0.050, "HTML": 0.025, "SMS": 0.020, "APP": 0.035}).fillna(0)
segmento_effect = _tmp["segmento_cliente"].map({"Masivo": 0.000, "Emergente": 0.015, "Premium": 0.030}).fillna(0)

p_venta = (
    0.015
    + 0.18 * _tmp["score_modelo"]
    + 0.06 * _tmp["score_digital"]
    + canal_effect
    + segmento_effect
).clip(0.01, 0.45)

target_venta = rng.binomial(1, p_venta)

campanas_raw["resultado"] = np.where(
    target_venta == 1,
    "VENTA",
    rng.choice(["NO_INTERESADO", "SIN_CONTACTO", "REBOTE", "NO_CALIFICA"], size=n_contactos, p=[0.45, 0.30, 0.15, 0.10])
)

campanas_raw["comentario"] = rng.choice(
    ["cliente interesado", "solicita llamada", "no contesta", "mail abierto", "sin respuesta", "numero invalido"],
    size=n_contactos
)

campanas_raw.head()
```

## 2.2 Inyectamos problemas de calidad

Esto nos permitirá practicar limpieza como se hace en un proyecto real.

```python
campanas_raw = campanas_raw.copy()

# Montos con formatos mixtos
idx_monto_texto = rng.choice(campanas_raw.index, size=140, replace=False)
campanas_raw["monto_ofertado"] = campanas_raw["monto_ofertado"].astype("object")
campanas_raw.loc[idx_monto_texto, "monto_ofertado"] = (
    "S/ " + campanas_raw.loc[idx_monto_texto, "monto_ofertado"].astype(str)
)

# Missing y valores inválidos
campanas_raw.loc[rng.choice(campanas_raw.index, size=55, replace=False), "monto_ofertado"] = None
campanas_raw.loc[rng.choice(campanas_raw.index, size=20, replace=False), "score_modelo"] = None
campanas_raw.loc[rng.choice(campanas_raw.index, size=15, replace=False), "canal"] = None
campanas_raw.loc[rng.choice(campanas_raw.index, size=8, replace=False), "monto_ofertado"] = "sin dato"

# Outliers y negativos
outlier_idx = rng.choice(campanas_raw.index, size=8, replace=False)
campanas_raw.loc[outlier_idx, "monto_ofertado"] = 250_000

neg_idx = rng.choice(campanas_raw.index, size=6, replace=False)
campanas_raw.loc[neg_idx, "monto_ofertado"] = -100

# Duplicados exactos
duplicados = campanas_raw.sample(25, random_state=SEED)
campanas_raw = pd.concat([campanas_raw, duplicados], ignore_index=True)

campanas_raw.shape
```

### Desglose de la lógica: Parquet vs. CSV

1. **El formato Parquet (`to_parquet` y `read_parquet`)**
   - **¿Qué es?**: Parquet es un formato de archivo de **columna** (columnar) optimizado diseñado para grandes volúmenes de datos.
   - **Ventajas sobre CSV**:
     - **Velocidad**: Es significativamente más rápido de leer y escribir (a veces 10x o más) porque no necesita analizar texto plano.
     - **Tamaño**: Comprime los datos mucho mejor, reduciendo drásticamente el espacio en disco.
     - **Tipado**: Guarda los tipos de datos (fechas, enteros, categorías) de forma nativa. **No necesitas** usar `parse_dates` o `dtype` al leerlo; pandas recuerda automáticamente cómo estaban guardados.
   - **Requisito**: Para funcionar, el entorno de Python necesita tener instalada la librería **`pyarrow`** o **`fastparquet`**.

## 4.2 Función de perfilado

La siguiente función resume lo mínimo que se revisa antes de transformar.

```python
def perfilar_dataframe(df: pd.DataFrame #data frame de entrada 
                       , nombre: str = "df"#Un parámetro opcional para etiquetar el reporte
                      ) -> pd.DataFrame: #Indica que la función devuelve un DataFrame con el resumen.
    # Devuelve un resumen práctico de estructura, nulos, cardinalidad y memoria.
    resumen = pd.DataFrame({
        "columna": df.columns,  #Lista los nombres de las columnas (df.columns).
        "dtype": df.dtypes.astype(str).values,  #Convierte los tipos de datos a texto (ej. int64, category, datetime).
        "nulos": df.isna().sum().values,  #Cuenta cuántos valores faltantes (NaN) hay en cada columna (df.isna().sum()).
        #Calcula el porcentaje de nulos:
            #df.isna().mean(): Da el promedio de nulos (proporción).
            #.mul(100): Lo convierte a porcentaje.
            #.round(2): Redondea a 2 decimales.
        "nulos_pct": df.isna().mean().mul(100).round(2).values,  
        "n_unicos": df.nunique(dropna=True).values, #Cuenta cuántos valores únicos hay en cada columna (cardinalidad)
        "memoria_mb": (df.memory_usage(deep=True) / 1e6).round(4).values[1:],
    })

    print(f"{nombre}: {df.shape[0]:,} filas x {df.shape[1]:,} columnas")
    print(f"Memoria total: {df.memory_usage(deep=True).sum() / 1e6:,.2f} MB")
    print(f"Duplicados exactos: {df.duplicated().sum():,}")

    return resumen.sort_values(["nulos_pct", "n_unicos"], ascending=[False, False])

perfil_campanas = perfilar_dataframe(campanas, "campanas")
perfil_campanas
```

### 📊 Análisis de Perfil de Datos (`perfilar_dataframe`)

Esta función automatiza la fase inicial de **Exploratory Data Analysis (EDA)** generando un informe detallado en una sola tabla.

#### ¿Qué calcula?
1.  **Estructura básica**: Muestra el tipo de dato (`dtype`) de cada columna.
2.  **Calidad de datos**:
    -   `nulos`: Cantidad exacta de valores faltantes (`NaN`).
    -   `nulos_pct`: Porcentaje de nulos (crucial para decidir si imputar o eliminar columnas).
3.  **Cardinalidad**:
    -   `n_unicos`: Número de valores distintos. Ayuda a identificar si una columna es categórica (pocos únicos) o de alta cardinalidad.
4.  **Eficiencia**:
    -   `memoria_mb`: Espacio que ocupa cada columna en memoria (ajustado para excluir el índice).

#### Estadísticas globales impresas en consola:
-   **Tamaño**: Filas y columnas totales.
-   **Memoria total**: Uso de RAM del objeto completo.
-   **Duplicados**: Filas idénticas que podrían estar sesgando el modelo.

#### Ordenamiento inteligente:
El resultado se ordena automáticamente para priorizar:
1.  Columnas con **mayor porcentaje de nulos** (puntos críticos de limpieza).
2.  Columnas con **mayor variedad de valores** (útiles para discriminación).

> 💡 **Tip**: Si ves columnas con `nulos_pct` > 50%, considera eliminarlas. Si ves columnas `object` con pocos `n_unicos`, conviértelas a `category` para ahorrar memoria.

Para esos usamos `.sort_values()`

## 4.3 Optimización de tipos

En bases grandes, elegir bien los `dtype` reduce memoria y evita errores posteriores.

Regla práctica:

- Categorías repetitivas: `category`.
- Fechas: `datetime64`.
- Montos y scores: numéricos.
- Enteros con nulos: tipos `Int64` o imputación previa.

```python
memoria_antes = campanas.memory_usage(deep=True).sum() / 1e6

campanas_opt = campanas.copy()
for col in ["canal", "producto", "resultado"]:
    campanas_opt[col] = campanas_opt[col].astype("category")

memoria_despues = campanas_opt.memory_usage(deep=True).sum() / 1e6

pd.DataFrame({
    "escenario": ["antes", "despues"],
    "memoria_mb": [memoria_antes, memoria_despues],
    "reduccion_pct": [0, 100 * (1 - memoria_despues / memoria_antes)]
})
```

## 5.2 Filtros booleanos con `.loc`

Pregunta aplicada:  
**¿Qué contactos de campañas de préstamo personal por canales digitales terminaron en venta?**

```python
canales_digitales = ["HTML", "html ", "SMS", "sms", "APP", "app "]

mask = (
    campanas["producto"].eq("PP") #.eq() es equivalente a == pero más seguro (maneja NaN de forma predecible)
    & campanas["canal"].isin(canales_digitales) #Filtra donde el canal está en la lista definida arriba (incluyendo las variaciones con espacios).
    & campanas["resultado"].eq("VENTA")
)

cols = ["contacto_id", "fecha_contacto", "cliente_id", "canal", "producto", "monto_ofertado", "score_modelo", "resultado"]

campanas.loc[mask, cols].head(10)
#campanas.loc[mask, cols]:
#mask: Selecciona solo las filas que cumplieron las condiciones.
#cols: Selecciona solo las columnas especificadas (eliminando ruido).
```

## 5.4 `.query()` para filtros expresivos

Primero convertimos `score_modelo` a numérico para poder filtrar.

Este código implementa una estrategia de limpieza y filtrado seguro en dos pasos. 
Su objetivo es filtrar datos basándose en una condición numérica `(score_modelo >= 0.60)` sin que el código falle si hay valores inválidos (texto o vacíos) en la columna original.



```python
campanas_query = campanas.assign(
    score_modelo_num=pd.to_numeric(campanas["score_modelo"], errors="coerce")
)
#campanas.assign(...): Crea una copia temporal del DataFrame campanas agregando una nueva columna llamada score_modelo_num sin modificar el original.
#Intenta convertir la columna score_modelo (que puede ser texto) a números.
#errors="coerce": Si encuentra un valor que no se puede convertir (ej. "N/A", "desconocido", o un espacio en blanco), no lanza un error. En su lugar, lo convierte en NaN.
campanas_query.query("producto == 'PP' and score_modelo_num >= 0.60").head() #filtro
```

## 5.5 MultiIndex para reportes jerárquicos

Cuando el resultado está naturalmente agrupado por varias claves, un `MultiIndex` facilita análisis por nivel.

```python
ventas_por_canal_producto = (
    campanas
    .assign(target_venta=lambda d: d["resultado"].eq("VENTA").astype(int))
    .groupby(["canal", "producto"])["target_venta"] #Divide los datos en grupos únicos basados en la combinación de canal y producto.Selecciona solo la columna target_venta para hacer los cálculos agregados.
    .agg(["count", "sum", "mean"]) #Calcula esas métricas
    .rename(columns={"count": "contactos", "sum": "ventas", "mean": "tasa_venta"}) #Renombra las columnas
    .sort_values("tasa_venta", ascending=False) 
)

ventas_por_canal_producto.head(10)
```

```python
# Acceso a un nivel del índice
ventas_por_canal_producto.xs("PP", level="producto").head()  #Filtra PP con respecto producto
```

## 6.3 Conversión robusta de montos

`monto_ofertado` llegó mezclado: números, texto, moneda, negativos y nulos.  
Primero limpiamos símbolos y luego convertimos con `pd.to_numeric`.

```python
def limpiar_monto(s: pd.Series) -> pd.Series:
    return (
        s.astype("string") #Convierte todo a string
        .str.replace(r"[^0-9.\-]", "", regex=True) #Reemplaza los Cualquier carácter que NO sea un dígito (0-9), un punto (.) o un guion (-) por ""
            #Usa una expresión regular (Regex) para limpiar el texto.
        .replace("", pd.NA) #reemplaza los "" por NA 
        .pipe(pd.to_numeric, errors="coerce") #Usa .pipe() para pasar la serie limpia a pd.to_numeric, Si queda algún valor que no se pueda convertir a número, lo fuerza a NaN.
    )

campanas_limpia["monto_ofertado_num"] = limpiar_monto(campanas_limpia["monto_ofertado"])
campanas_limpia["score_modelo_num"] = pd.to_numeric(campanas_limpia["score_modelo"], errors="coerce")

campanas_limpia[["monto_ofertado", "monto_ofertado_num", "score_modelo", "score_modelo_num"]].head(10)
```

```python
campanas_limpia[campanas_limpia.cliente_id == 100262]
```

## 6.4 Missing data

Tratamos nulos según el significado:

- `canal`: categoría desconocida.
- `score_modelo`: mediana.
- `monto_ofertado`: mediana por producto.

```python
campanas_limpia["canal"] = campanas_limpia["canal"].fillna("DESCONOCIDO")

campanas_limpia["score_modelo_num"] = (
    campanas_limpia["score_modelo_num"]
    .fillna(campanas_limpia["score_modelo_num"].median())
    .clip(0, 1)
)

mediana_monto_producto = campanas_limpia.groupby("producto")["monto_ofertado_num"].transform("median")

campanas_limpia["monto_ofertado_num"] = (
    campanas_limpia["monto_ofertado_num"]
    .fillna(mediana_monto_producto)
    .fillna(campanas_limpia["monto_ofertado_num"].median())
)

campanas_limpia[["canal", "score_modelo_num", "monto_ofertado_num"]].isna().sum()
```

## 6.5 Outliers con `clip`

Los montos extremos pueden distorsionar promedios, KPIs y modelos.  
Usaremos winsorización al percentil 1 y 99.

```python
q01, q99 = campanas_limpia["monto_ofertado_num"].quantile([0.01, 0.99])

campanas_limpia["monto_ofertado_clip"] = (
    campanas_limpia["monto_ofertado_num"]
    .clip(lower=max(q01, 0), upper=q99)
)

campanas_limpia[["monto_ofertado_num", "monto_ofertado_clip"]].describe().T
```

## 6.6 Feature engineering con `.assign()`

Creamos variables útiles para análisis comercial y Data Science:

- `target_venta`
- `mes`
- `dia_semana`
- `canal_digital`
- `costo_contacto`
- `ingreso_esperado`
- `roi_esperado`
- segmentos de score y monto

```python
costo_por_canal = {
    "TLV": 2.20,
    "HTML": 0.04,
    "SMS": 0.12,
    "APP": 0.02,
    "DESCONOCIDO": 0.20,
}

campanas_fe = (
    campanas_limpia
    .assign(
        target_venta=lambda d: d["resultado"].eq("VENTA").astype(int),
        mes=lambda d: d["fecha_contacto"].dt.to_period("M").astype(str),
        dia_semana=lambda d: d["fecha_contacto"].dt.day_name(),
        canal_digital=lambda d: d["canal"].isin(["HTML", "SMS", "APP"]).astype(int),
        costo_contacto=lambda d: d["canal"].map(costo_por_canal).fillna(0.20),
        ingreso_esperado=lambda d: d["score_modelo_num"] * d["monto_ofertado_clip"],
        roi_esperado=lambda d: d["ingreso_esperado"] - d["costo_contacto"],
        score_seg=lambda d: pd.cut(
            d["score_modelo_num"],
            bins=[-0.01, 0.20, 0.45, 1.00],
            labels=["bajo", "medio", "alto"],
        ),
        monto_seg=lambda d: pd.qcut(
            d["monto_ofertado_clip"],
            q=4,
            labels=["Q1", "Q2", "Q3", "Q4"],
            duplicates="drop",
        ),
    )
)

campanas_fe.head()
```

## 6.7 `np.where` y `np.select`

Creamos una regla de priorización comercial:

- `contactar_tlv`: alto score y monto alto.
- `digital_barato`: buen score, pero no justifica canal costoso.
- `nutrir`: score bajo o intermedio.

```python
condiciones = [
    (campanas_fe["score_modelo_num"] >= 0.45) & (campanas_fe["monto_seg"].isin(["Q3", "Q4"])),
    (campanas_fe["score_modelo_num"] >= 0.25) & (campanas_fe["canal_digital"].eq(1)),
    (campanas_fe["score_modelo_num"] < 0.25),
]

decisiones = ["contactar_tlv", "digital_barato", "nutrir"]

campanas_fe["estrategia"] = np.select(condiciones, decisiones, default="revisar")
#Es booleano, el default es por si no se cumplen

campanas_fe["estrategia"].value_counts()
#campanas_fe.head()
```

## 6.8 Vectorización vs. `apply`

La mayoría de transformaciones debe hacerse de forma vectorizada.  
`apply(axis=1)` se reserva para casos donde la regla depende de muchas columnas y no existe alternativa simple.

```python
# Vectorizado: recomendado
campanas_fe["monto_ajustado_vectorizado"] = np.where(
    campanas_fe["producto"].eq("PP"), #Condición
    campanas_fe["monto_ofertado_clip"] * 1.05, #Valor_si_veradero
    campanas_fe["monto_ofertado_clip"], #Valor_si_falso
)

# apply: más flexible, pero menos eficiente
def ajustar_monto(row):
    if row["producto"] == "PP":
        return row["monto_ofertado_clip"] * 1.05
    return row["monto_ofertado_clip"]

campanas_fe["monto_ajustado_apply"] = campanas_fe.apply(ajustar_monto, axis=1)

np.allclose(campanas_fe["monto_ajustado_vectorizado"], campanas_fe["monto_ajustado_apply"])
```

# 10. Series temporales: resample, rolling, shift y pct_change

Cuando el índice es una fecha, Pandas permite reagregar por día, semana, mes y calcular ventanas móviles.

## 10.1 Revenue esperado diario y semanal

```python
ts = (
    campanas_clientes
    .set_index("fecha_contacto") #Convertir la columna fecha_contacto en el índice del DataFrame y ordenarlo cronológicamente.
    .sort_index()
)

daily = ts["ingreso_esperado"].resample("D").sum() #Agrupa los datos por día
weekly = ts["ingreso_esperado"].resample("W").sum() #Agrupa los datos por semanas

pd.DataFrame({
    "daily_head": daily.head(7),
})
```

```python
weekly.head(10)
```

## 10.2 Media móvil y crecimiento

La media móvil suaviza la serie para ver tendencia.  
`shift`, `diff` y `pct_change` ayudan a crear lags y variaciones.

```python
#Convierte el objeto daily (probablemente una Serie de pandas) en un DataFrame y le asigna la columna ingreso_esperado.
serie_diaria = daily.to_frame("ingreso_esperado")
#Media Móvil de 7 días. Suaviza las fluctuaciones diarias para revelar la tendencia subyacente. El argumento min_periods=1 asegura que la primera fila tenga un valor (aunque sea solo el promedio del primer día) en lugar de NaN.
serie_diaria["ma7"] = serie_diaria["ingreso_esperado"].rolling(7, min_periods=1).mean()
#Valor con retraso de 1 día. Empuja los datos un paso hacia abajo. El valor de hoy es el de ayer. Es crucial para crear modelos predictivos donde el pasado predice el futuro.
serie_diaria["lag1"] = serie_diaria["ingreso_esperado"].shift(1)
serie_diaria["diff1"] = serie_diaria["ingreso_esperado"].diff()
serie_diaria["pct_change1"] = serie_diaria["ingreso_esperado"].pct_change()

serie_diaria.head(10)
```

# 11. Caso integrador: pipeline final

Construiremos un pipeline reproducible que:

1. Limpia campañas.
2. Enriquece con clientes.
3. Crea features.
4. Genera KPIs.
5. Construye una tabla agregada para modelamiento.

## 11.1 Función de limpieza y feature engineering

```python
def preparar_campanas(df_raw: pd.DataFrame) -> pd.DataFrame:
    # Limpia campañas y genera variables analíticas.
    costo_por_canal = {
        "TLV": 2.20,
        "HTML": 0.04,
        "SMS": 0.12,
        "APP": 0.02,
        "DESCONOCIDO": 0.20,
    }

    df = df_raw.copy()

    df = (
        df
        .drop_duplicates(subset=["contacto_id"], keep="last")
        .assign(
            fecha_contacto=lambda d: pd.to_datetime(d["fecha_contacto"], errors="coerce"),
            canal=lambda d: d["canal"].astype("string").str.strip().str.upper().fillna("DESCONOCIDO"),
            producto=lambda d: d["producto"].astype("string").str.strip().str.upper(),
            monto_ofertado_num=lambda d: limpiar_monto(d["monto_ofertado"]),
            score_modelo_num=lambda d: pd.to_numeric(d["score_modelo"], errors="coerce"),
            target_venta=lambda d: d["resultado"].eq("VENTA").astype(int),
        )
    )

    mediana_monto_producto = df.groupby("producto")["monto_ofertado_num"].transform("median")

    df = (
        df
        .assign(
            monto_ofertado_num=lambda d: d["monto_ofertado_num"].fillna(mediana_monto_producto),
            score_modelo_num=lambda d: d["score_modelo_num"].fillna(d["score_modelo_num"].median()).clip(0, 1),
        )
    )

    q01, q99 = df["monto_ofertado_num"].quantile([0.01, 0.99])

    df = (
        df
        .assign(
            monto_ofertado_clip=lambda d: d["monto_ofertado_num"].fillna(d["monto_ofertado_num"].median()).clip(max(q01, 0), q99),
            mes=lambda d: d["fecha_contacto"].dt.to_period("M").astype(str),
            costo_contacto=lambda d: d["canal"].map(costo_por_canal).fillna(0.20),
            canal_digital=lambda d: d["canal"].isin(["HTML", "SMS", "APP"]).astype(int),
            ingreso_esperado=lambda d: d["score_modelo_num"] * d["monto_ofertado_clip"],
            roi_esperado=lambda d: d["ingreso_esperado"] - d["costo_contacto"],
            score_seg=lambda d: pd.cut(d["score_modelo_num"], [-0.01, 0.20, 0.45, 1.00], labels=["bajo", "medio", "alto"]),
        )
    )

    return df


campanas_pipeline = preparar_campanas(campanas)
campanas_pipeline.head()
```

## 11.2 Enriquecimiento con clientes

```python
base_analitica = campanas_pipeline.merge(
    clientes_base,
    on="cliente_id",
    how="left",
    validate="many_to_one",
    indicator=True,
)

base_analitica["_merge"].value_counts()
```

## 11.3 KPIs ejecutivos

Estos KPIs pueden alimentar una lámina, dashboard o comité comercial.

```python
kpis_ejecutivos = (
    base_analitica
    .groupby(["mes", "canal", "producto"])
    .agg(
        contactos=("contacto_id", "count"),
        clientes=("cliente_id", "nunique"),
        ventas=("target_venta", "sum"),
        tasa_venta=("target_venta", "mean"),
        monto_promedio=("monto_ofertado_clip", "mean"),
        costo_total=("costo_contacto", "sum"),
        ingreso_esperado=("ingreso_esperado", "sum"),
        roi_esperado=("roi_esperado", "sum"),
    )
    .assign(
        costo_por_contacto=lambda d: d["costo_total"] / d["contactos"],
        costo_por_venta=lambda d: d["costo_total"] / d["ventas"].replace(0, np.nan),
    )
    .round(4)
    .sort_values(["mes", "roi_esperado"], ascending=[True, False])
)

kpis_ejecutivos.head(15)
```

## 11.4 Tabla para modelamiento

Unidad de análisis: cliente-mes.

Target: si el cliente tuvo al menos una venta en el mes.  
Features: intensidad de contactos, score promedio, monto promedio, canales usados y variables del cliente.

```python
tabla_modelo = (
    base_analitica
    .groupby(["cliente_id", "mes"])
    .agg(
        target_venta_mes=("target_venta", "max"),
        contactos_mes=("contacto_id", "count"),
        score_promedio=("score_modelo_num", "mean"),
        score_max=("score_modelo_num", "max"),
        monto_promedio=("monto_ofertado_clip", "mean"),
        monto_max=("monto_ofertado_clip", "max"),
        roi_promedio=("roi_esperado", "mean"),
        canales_distintos=("canal", "nunique"),
        productos_distintos=("producto", "nunique"),
        share_digital=("canal_digital", "mean"),
        edad=("edad", "first"),
        ingreso_mensual=("ingreso_mensual", "first"),
        score_digital=("score_digital", "first"),
        antiguedad_meses=("antiguedad_meses", "first"),
        segmento_cliente=("segmento_cliente", "first"),
        region=("region", "first"),
    )
    .reset_index()
)

tabla_modelo.head()
```

```python
tabla_modelo["target_venta_mes"].value_counts(normalize=True).rename("distribucion_target")
```

# 13. Cierre

Ideas clave:

1. Pandas no es solo sintaxis: es el lenguaje operativo para convertir datos crudos en decisiones.
2. Antes de transformar, siempre se perfila.
3. Los tipos correctos reducen memoria y previenen errores.
4. `.loc` evita asignaciones ambiguas.
5. La mayoría de features se crean con operaciones vectorizadas.
6. `groupby`, `merge`, `pivot_table`, `resample` y `rolling` cubren gran parte del trabajo real.
7. Todo análisis serio debe terminar en un pipeline reproducible.
