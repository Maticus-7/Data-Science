## Ejercicio 1.1 — Clasificación de Cartera por Tamaño de Préstamo *(7 pts)*

**Contexto:** El equipo de riesgos de BancoSur recibe cada semana un lote de nuevas solicitudes de crédito. Para enrutarlas al circuito de aprobación correcto, primero se clasifican por tamaño: los créditos **Micro** los aprueba el sistema automático, los **Pequeños** pasan por un analista junior, los **Medianos** requieren comité, y los **Grandes** necesitan aprobación del directorio.

| Categoría | Criterio |
|-----------|----------|
| Micro     | monto < 5 000 |
| Pequeño   | 5 000 ≤ monto < 20 000 |
| Mediano   | 20 000 ≤ monto < 50 000 |
| Grande    | monto ≥ 50 000 |

**Se pide:**
1. Recorrer `montos_solicitudes` con un bucle `for`
2. Clasificar cada monto con `if / elif / else`
3. Mantener **contadores** y **acumuladores de suma** por categoría
4. Guardar los montos individuales de cada categoría en **listas** usando `append`
5. Al finalizar, imprimir una tabla con: Categoría | Cantidad | Suma Total | Promedio

```python
# ============================================================
# EJERCICIO 1.1 — Clasificación de Cartera por Tamaño
# ============================================================

# --- DATOS DADOS — no modificar ---
montos_solicitudes = [
    3500, 12000, 8750, 67000, 4200, 25000, 45000, 1800,
    31500, 82000, 9500, 19800, 55000, 2900, 14000,
    38500, 4800, 23000, 71000, 6500
]
print("-"*100)
# ── PASO 1: Inicializar acumuladores ──────────────────────
print("# ── PASO 1: Inicializar acumuladores ──────────────────────")
# Para cada categoría necesitas: un contador (int),
# una suma (float/int) y una lista de montos individuales.

# Contadores (int)
contador_micro = 0
contador_pequeño = 0
contador_mediano = 0
contador_grande = 0

# Sumas (float/int)
sum_micro = 0.0
sum_pequeño = 0.0
sum_mediano = 0.0
sum_grande = 0.0

# Listas de montos individuales
list_micro = []
list_pequeño = []
list_mediano = []
list_grande = []

print("-"*100)
# ── PASO 2: Bucle de clasificación ───────────────────────
print("# ── PASO 2: Bucle de clasificación ───────────────────────")
for monto in montos_solicitudes:
    #     determinar a qué categoría pertenece
    #     actualizar: contador, suma y lista
    if monto < 5000:
        contador_micro += 1
        sum_micro += monto
        list_micro.append(monto)
    elif monto < 20000:
        contador_pequeño += 1
        sum_pequeño += monto
        list_pequeño.append(monto)
    elif monto < 50000:
        contador_mediano += 1
        sum_mediano += monto
        list_mediano.append(monto)
    else:
        contador_grande += 1
        sum_grande += monto
        list_grande.append(monto)

print("-"*100)
# ── PASO 3: Imprimir tabla resumen ────────────────────────
print("# ── PASO 3: Imprimir tabla resumen ────────────────────────\n")
# Calcular promedio = suma / contador
def calcular_promedio(total, cantidad):
    return total / cantidad if cantidad > 0 else 0.0
# Imprimir una fila por categoría y fila TOTAL al final
print(f"{'Categoria':<12} | {'Cantidad':<12} | {'Total':<12} | {'Promedio':<12}") 
print("-" * 60)

categorias = [
    ("Micro", contador_micro, sum_micro),
    ("Pequeno", contador_pequeño, sum_pequeño),
    ("Mediano", contador_mediano, sum_mediano),
    ("Grande", contador_grande, sum_grande)
]

total_cantidad = 0 
total_suma = 0

for nombre, cantidad, total in categorias:
    promedio = calcular_promedio(total, cantidad)
    print(f"{nombre:<12} | {cantidad:<12} | {total:<12.2f} | {promedio:<12.2f}")
    total_cantidad += cantidad
    total_suma += total

print("-" * 60)
print(f"{'TOTAL':<12} | {total_cantidad:<12} | {total_suma:<12.2f} | {calcular_promedio(total_suma, total_cantidad):<12.2f}")
```

## Ejercicio 1.2 — Sistema de Evaluación Crediticia *(8 pts)*

**Contexto:** BancoSur quiere automatizar la evaluación inicial de solicitudes de crédito. Te piden construir la función `evaluar_credito` que, dado el ingreso mensual, la deuda mensual y el score crediticio de un cliente, devuelva una decisión estructurada.

**Reglas de negocio (en orden de prioridad):**

| Condición | Decisión |
|-----------|----------|
| `ratio > 0.40` *(deuda > 40% del ingreso)* | **Rechazado** |
| `score < 600` *(score mínimo no alcanzado)* | **Rechazado** |
| `600 ≤ score ≤ 699` **Y** `ratio > 0.30` | **Revisar** *(comité manual)* |
| Cualquier otro caso | **Aprobado** |

> Las condiciones de rechazo tienen prioridad. Si `ratio > 0.40`, se rechaza aunque el score sea excelente.

**Se pide:**
1. Definir `evaluar_credito(ingreso, deuda, score)` que calcule el `ratio_endeudamiento`, aplique las reglas en orden, y **retorne un diccionario** con claves: `ratio_endeudamiento`, `decision`, `motivo`
2. Llamar a la función para **cada cliente** de la lista
3. Imprimir: nombre, ratio (2 decimales), decisión y motivo de forma legible

```python
# ============================================================
# EJERCICIO 1.2 — Sistema de Evaluación Crediticia
# ============================================================

# --- DATOS DADOS — no modificar ---
clientes = [
    {"nombre": "Ana Torres",   "ingreso": 5_500,  "deuda": 1_800, "score": 720},
    {"nombre": "Luis Paredes", "ingreso": 3_200,  "deuda": 1_400, "score": 580},
    {"nombre": "María Quispe", "ingreso": 8_000,  "deuda": 2_800, "score": 645},
    {"nombre": "Carlos Vera",  "ingreso": 4_500,  "deuda": 1_950, "score": 630},
    {"nombre": "Rosa Mamani",  "ingreso": 12_000, "deuda": 2_000, "score": 710},
]

# ── PASO 1: Definir la función ────────────────────────────
def evaluar_credito(ingreso, deuda, score):
    #Calculamos el ratio:
     ratio = deuda / ingreso if ingreso > 0 else 1
    #Colocamos 1 pues , si el ingreso es 0 , eso quiere decir que el ratio es un número infinitamente grande,
    #Como el máximo valor es 1 (al ser normado) ,entonces basta tomar un número mayor o igual a 1.
    #Las Reglas:
     if ratio > 0.40:
         decision = "Rechazado"
         motivo   = "Deuda > 40% del ingreso"
     elif score < 600:
         decision = "Rechazado"
         motivo   = "Score mínimo no alcanzado"
     elif (score < 699) and (ratio > 0.30):
         decision = "Revisar"
         motivo   = "Comité manual"
     else:
         decision = "Aprobado"
         motivo = "Cumple todos los criterios"

    #Retornamos el diccionario:
     return {
        "decision": decision,
        "motivo": motivo,
        "ratio": round(ratio,4),
        "score": score
    }

# ── PASO 2: Aplicar a cada cliente e imprimir ────────────
print(f"{'Cliente':<15} | {'Ratio':<8} | {'Decisión':<10} | {'Motivo'}")
print("-" * 80)

for cliente in clientes:
    nombre = cliente["nombre"]
    ingreso = cliente["ingreso"]
    deuda = cliente["deuda"]
    score = cliente["score"]
    
    resultado = evaluar_credito(ingreso, deuda, score)
#     print(...)
    print(f"{nombre:<15} | {resultado['ratio']:<8.2f} | {resultado['decision']:<10} | {resultado['motivo']}")
```

## Ejercicio 1.3 — Análisis de Productos con Sets y Comprensiones *(7 pts)*

**Contexto:** El equipo de cross-selling de BancoSur quiere entender el comportamiento de solicitudes de producto por cliente.

**Se pide:**
1. Usando un **set** (o set comprehension), obtener todos los **productos únicos** del catálogo
2. Usando una **comprensión de lista**, obtener los IDs de clientes que solicitaron `"Préstamo Personal"` *(puede haber duplicados)*
3. Usando un diccionario `{cliente: set_de_productos}`, identificar qué clientes solicitaron **más de un producto distinto** — imprimir su ID y los productos
4. Imprimir cada resultado con un mensaje descriptivo

```python
# ============================================================
# EJERCICIO 1.3 — Sets y Comprensiones de Lista
# ============================================================

# --- DATOS DADOS — no modificar ---
solicitudes = [
    ["C001", "Préstamo Personal"],
    ["C002", "Tarjeta de Crédito"],
    ["C003", "Préstamo Personal"],
    ["C004", "Hipoteca"],
    ["C005", "Préstamo Vehicular"],
    ["C002", "Préstamo Personal"],
    ["C006", "Tarjeta de Crédito"],
    ["C007", "Hipoteca"],
    ["C001", "Tarjeta de Crédito"],
    ["C008", "Préstamo Personal"],
    ["C003", "Préstamo Vehicular"],
    ["C009", "Hipoteca"],
]

# PASO 1: Productos únicos con set comprehension
print("."*100)
print("# PASO 1: Productos únicos con set comprehension")
# Usamos las {} para crear un set y eliminando duplicados.
productos_unicos = {producto for _, producto in solicitudes} #Con _, le digo a python que solo me interesa el segundo valor.

print("1. Catálogo de Productos:")
print(f"   {productos_unicos}")
print("-" * 100)

# PASO 2: Clientes que pidieron "Préstamo Personal" (list comprehension)
print("# PASO 2: Clientes que pidieron 'Préstamo Personal' (list comprehension)")
# No usamos set porque puede haber duplicado
clientes_pp = [cliente for cliente, producto in solicitudes if producto == "Préstamo Personal"]

print("2. Clientes que solicitaron 'Préstamo Personal':")
print(f"   {clientes_pp}")
print("-" * 100)

# PASO 3: Clientes con más de un producto distinto
print("# PASO 3: Clientes con más de un producto distinto")
# Construir dict {cliente: set_de_productos}, luego filtrar
print("# 3a. Construir el diccionario {cliente: set_de_productos}")
productos_por_cliente = {}
for cliente, producto in solicitudes:
    if cliente not in productos_por_cliente:
        productos_por_cliente[cliente] = set()
    productos_por_cliente[cliente].add(producto)
    
for cliente,producto in productos_por_cliente.items():
    print(f"Cliente {cliente} , Productos: {producto}" )

print("-" * 100)

print("# 3b. Filtrar para obtener solo los clientes con > 1 producto")
# Usamos dict comprehension para crear el nuevo diccionario filtrado.
clientes_multi = {
    cliente: productos 
    for cliente, productos in productos_por_cliente.items() 
    if len(productos) > 1
}

print("3. Clientes con más de un producto distinto:")
if not clientes_multi:
    print("No hay clientes con más de un producto.")
else:
    for cliente, productos in clientes_multi.items():
        # Formateamos la salida para que sea legible
        productos_str = ", ".join(sorted(productos))
        print(f"   - Cliente {cliente}: {productos_str}")

print("-" * 100)
```

## Ejercicio 1.4 — Simulación de Tabla de Amortización *(8 pts)*

**Contexto:** Un analista de BancoSur necesita calcular cuánto tiempo tardará un cliente en pagar su deuda y cuánto pagará en total, dada una cuota mensual fija y una tasa de interés sobre el saldo pendiente.

**Mecánica del pago mensual:**
1. `interes_mes = saldo_actual × tasa_mensual`
2. `abono_capital = cuota_mensual − interes_mes`
3. `saldo_actual = saldo_actual − abono_capital`
4. Repetir hasta que `saldo_actual ≤ 0`

> **Caso del último mes:** si el saldo restante es menor que el abono de capital, el cliente solo paga lo que queda más los intereses de ese mes.

**Se pide:**
1. Usar un bucle `while` que simule los pagos mes a mes
2. Imprimir una tabla con los **primeros 6 meses**, `"..."`, y el **último mes** — columnas: Mes | Cuota | Interés | Abono Capital | Saldo Pendiente
3. Al finalizar, imprimir en resumen: número de meses, total pagado, total intereses, costo financiero como % del capital

```python
# ============================================================
# EJERCICIO 1.4 — Tabla de Amortización
# ============================================================

# --- DATOS DADOS — no modificar ---
saldo_actual   = 25_000.0   # S/. de deuda total (variable)
capital = 25_000.0  # S/. de deuda total (fijo)
cuota_mensual  = 700.0      # S/. que el cliente paga cada mes
tasa_mensual   = 0.015      # 1.5% mensual sobre el saldo pendiente

mes          = 0
total_pagado = 0.0
registros    = []  # guardar cada fila (mes, cuota, interes, abono, saldo)

# ── PASO 1: Bucle while ───────────────────────────────────
while saldo_actual > 0:
    mes += 1
    #Calcular interes_mes, abono_capital
    interes_mes = saldo_actual * tasa_mensual
    abono_capital = cuota_mensual - interes_mes
    #Manejo del último mes (si saldo < abono normal)
    if saldo_actual < abono_capital:
        cuota_final = saldo_actual + interes_mes #La ultima cuota
        saldo_actual = 0 #En Teoría el saldo debería acabar siendo negativo pero para beneficio de la empresa acaba siendo 0
        total_pagado += cuota_final #Sería el total final pagado
        registros.append([mes, cuota_final, interes_mes, abono_capital, saldo_actual])
        break
    #Actualizar saldo_actual y total_pagado
    saldo_actual = saldo_actual - abono_capital
    total_pagado += cuota_mensual #No puede ser el abono ,pues eso es lo q estamos reduciendo 
    registros.append([mes, cuota_mensual, interes_mes, abono_capital, saldo_actual])   # guardar datos del mes


# ── PASO 2: Imprimir tabla (primeros 6 + "..." + último) ──
print(f"{'Mes':<7} | {'Cuota':<10} | {'Interés':>20} | {'Abono Capital': >20}| {'Saldo Pendiente':>20}")
print("-" * 100)

for r in registros:
    if r[0]<=6:
        print(f"{r[0]:<7} | {r[1]:<10} | {r[2]:>20} | {r[3]:>20} | {r[4]:>20} ")

for _ in range(3):
    print(".")

r_u = registros[-1]
print(f"{r_u[0]:<7} | {r_u[1]:<10} | {r_u[2]:>20} | {r_u[3]:>20} | {r_u[4]:>20} ")

# ── PASO 3: Imprimir resumen ──────────────────────────────
#número de meses, total pagado, total intereses, costo financiero como % del capital
print("\n En resumen:")
print("Número de meses:",mes)
print("Total pagado:",total_pagado)
print("Total intereses:",sum([r[2] for r in registros]))
print(f"Costo financiero: {(total_pagado - capital) / capital}%" )
```

   - `'Medio'` si 600 ≤ score < 700
   - `'Bajo'` si score ≥ 700

4. **`categoria_monto`** basado en `monto` (mismas reglas del Ej. 1.1: Micro/Pequeño/Mediano/Grande), usando `apply` con una **función definida** (`def`) o `lambda`

## Ejercicio 3.1 — Distribución de Poisson: Equipo de Alertas *(7 pts)*

**Contexto:** El sistema de monitoreo de fraudes de BancoSur genera en promedio **λ = 3.8 alertas por hora**. Las alertas llegan de forma independiente.

**Se pide:**
1. Calcular `P(X = 5)` — probabilidad de recibir exactamente 5 alertas en una hora
2. Calcular `P(X > 6)` — probabilidad de recibir más de 6 alertas *(usa 1 − CDF)*
3. ¿Cuántas alertas cubre el **90% de las horas**? Usar `ppf(0.90)` e interpretar: si el equipo tiene esa capacidad, ¿qué % de horas podrá atender toda la carga?
4. **Simulación:** generar 240 horas (30 días × 8h), calcular media y varianza de la simulación y comparar con los valores teóricos. Graficar distribución simulada vs. PMF teórica.

> `from scipy import stats` → `dist = stats.poisson(mu=3.8)`

```python
# ============================================================
# EJERCICIO 3.1 — Distribución de Poisson
# ============================================================
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

lam  = 3.8
dist = stats.poisson(mu=lam)
print("-"*100)
# 1. P(X = 5)
p_exactamente_5 = dist.pmf(5)
print(f"La probabilidad de recibir exactamente 5 alertas en una hora es del: {p_exactamente_5*100:.4f}%")
print("-"*100)

# 2. P(X > 6) = 1 - P(X <= 6)
p_mas_de_6 = 1 - dist.cdf(6)
print(f"La probabilidad de recibir más de 6 alertas en un hora: {p_mas_de_6*100:.4f}%")
print("-"*100)

# 3. Capacidad para 90% de las horas
capacidad_90 = dist.ppf(0.90)
print(f"Una capacidad de {int(capacidad_90)} alertas cubre exactamente el {dist.cdf(capacidad_90):.2%} de las horas.")
print("-"*100)

# 4. Simulación de 240 horas
np.random.seed(42)
simulacion = dist.rvs(size=240) #Simulamos
# Comparar media y varianza con teóricas
media_sim = np.mean(simulacion) #Calcula la media de la simulación
var_sim = np.var(simulacion, ddof=0) #Calcula la varianza de la simulación

print(f"\nMedia Teórica: {lam} | Media Simulada: {media_sim:.4f}")
print(f"Varianza Teórica: {lam} | Varianza Simulada: {var_sim:.4f}")
# Graficar simulación vs PMF teórica
k_vals = np.arange(0, max(simulacion) + 2) #Creamos la partición del eje X
pmf_vals = dist.pmf(k_vals) #Nos da la altura que debería tener cada punto teóricamente

plt.figure(figsize=(8, 5))
plt.hist(
    simulacion, 
    bins=np.arange(-0.5, max(simulacion) + 1.5, 1), #Dibujamos las barras que empiezen en -0.5 y terminen en 10+1.5 y de ancho 1
    density=True, 
    alpha=0.5, 
    color='steelblue', 
    edgecolor='black', 
    label='Simulación (240h)'
)

# Línea de probabilidad teórica
plt.plot(k_vals, pmf_vals, 'ro-', ms=6, label='PMF Teórica Poisson(3.8)')

plt.title('Simulación de Alertas vs PMF Teórica')
plt.xlabel('Número de Alertas por Hora')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.tight_layout()
plt.show()
```

## Ejercicio 3.2 — Distribución Normal: Zonas de Riesgo de Score *(7 pts)*

**Contexto:** El equipo de modelos de BancoSur ha determinado que el score crediticio del segmento Retail sigue una distribución Normal con **μ = 650** y **σ = 85**.

**Se pide:**
1. Calcular `P(score < 600)` — proporción de clientes en zona de riesgo alto
2. Calcular `P(600 ≤ score ≤ 750)` — proporción en zona aceptable
3. ¿Cuál es el score mínimo para pertenecer al **top 10%** de clientes? Usar `ppf(0.90)`
4. **Graficar** la curva Normal marcando visualmente tres zonas de riesgo:
   - 🔴 Roja: score < 600
   - 🟡 Naranja: 600 ≤ score < 700  
   - 🟢 Verde: score ≥ 700  
   - Añadir una línea punteada en el umbral del top 10%

> `dist = stats.norm(loc=650, scale=85)`

```python
# ============================================================
# EJERCICIO 3.2 — Distribución Normal
# ============================================================
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

mu    = 650
sigma = 85
dist  = stats.norm(loc=mu, scale=sigma)

# 1. P(score < 600) 
print("-"*100)
#Como la distribución es Normal ,entonces P(score < 600) = P(score <= 600)
p_bajo = dist.cdf(600) 
print(f"La proporción de clientes que caen en la zona de riesgo alto es: {p_bajo*100:.2f}%")
print("-"*100)

# 2. P(600 ≤ score ≤ 750) = P(score ≤ 750) -  P(600 < score) = P(score ≤ 750) -  P(600 ≤ score)
p_aceptable = dist.cdf(750) - dist.cdf(600)
print(f"La proporción de clientes que caen en la zona aceptables es: {p_aceptable*100:.2f}%")
print("-"*100)

# 3. Score mínimo top 10%
score_top10 = dist.ppf(0.90)
print(f"El score mínimo para pertenecer al top 10% de clientes es: {score_top10}")
print("-"*100)


# 4. Gráfico con zonas de color
print("# 4. Gráfico con zonas de color \n")
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 500)
y = dist.pdf(x)

fig, ax = plt.subplots(figsize=(11, 5)) #Tamaño del lienzo
ax.plot(x, y, color = 'black' , linewidth = 1.5 , label = f'N(μ={mu}, σ={sigma})')
#Rellenamos la zonas
ax.fill_between(x, y ,where=(x < 600), color = 'red', alpha = 0.6, label='Riesgo Alto (<600)') #Riesgo Alto
ax.fill_between(x, y ,where=((x >= 600) & (x < 700)) , color = 'yellow', alpha = 0.6, label='Riesgo Medio (600-699)') #Riesgo Medio
ax.fill_between(x, y ,where=(x >= 700), color = 'green', alpha = 0.6, label='Riesgo Bajo (>=700)') #Riesgo Alto
#Línea punteada en el umbral del top 10%
ax.axvline(score_top10, color='blue', linestyle='--', linewidth=2, label=f'Top 10% (Score >= {score_top10:.1f})')
#Títulos
ax.set_title('Distribución de Scores Crediticios(Zona de Riesgo) - Segmento Retail')
ax.set_xlabel('Score Crediticio')
ax.set_ylabel('Densidad de Probabilidad')
ax.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

| p < 0.001 | Muy fuerte | Rechazar H₀ |
| 0.001–0.01 | Fuerte | Rechazar H₀ |
| 0.01–0.05 | Significativa | Rechazar H₀ |
| 0.05–0.10 | Marginal | NO rechazar |
| p ≥ 0.10 | Compatible con H₀ | NO rechazar |

```python
# ============================================================
# EJERCICIO 3.3 — Z-Test de Dos Proporciones
# ============================================================
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

# --- DATOS DADOS — no modificar ---
n_S1, x_S1 = 10_200, 480   # Semestre 1
n_S2, x_S2 =  9_800, 412   # Semestre 2
print("-"*100)
# 1. Tasas observadas
p_S1 = x_S1 / n_S1  # Semestre 1
p_S2 = x_S2 / n_S2  # Semestre 2
print(f"1. Tasa S1: {p_S1:.4%} | Tasa S2: {p_S2:.4%}")
print("-"*100)
# 2. Z-test — H1: p_S2 < p_S1
z_stat, p_valor = proportions_ztest(
    count = [x_S2 , x_S1],
    nobs  = [n_S2 , n_S1],
    alternative = 'smaller'
)
print("-"*100)
print(f"2. Estadístico Z: {z_stat:.4f} | p-valor: {p_valor:.4f}")
print("-"*100)
# 3. Interpretación con rejilla operativa
if p_valor < 0.001:
    evidencia, decision = "Muy fuerte", "Rechazar H0"
elif p_valor <= 0.01:
    evidencia, decision = "Fuerte", "Rechazar H0"
elif p_valor <= 0.05:
    evidencia, decision = "Significativa", "Rechazar H0"
elif p_valor <= 0.10:
    evidencia, decision = "Marginal", "NO rechazar H0"
else:
    evidencia, decision = "Compatible con H0", "NO rechazar H0"

print("3. Interpretación (α=0.05):")
print(f"   Evidencia: {evidencia}")
print(f"   Decisión:  {decision}")
# 4. Conclusión de negocio (escríbela en la celda Markdown siguiente)
```

## Ejercicio 3.4 — ANOVA + Tukey: Score por Segmento *(8 pts)*

**Contexto:** La directora de análisis quiere saber si el `score_crediticio` promedio es estadísticamente distinto entre los tres segmentos del banco.

**Hipótesis:**
- **H₀:** μ_Retail = μ_PYME = μ_Corporativo
- **H₁:** Al menos un par de segmentos tiene medias distintas
- **α = 0.05**

**Se pide:**
1. Cargar `cartera_crediticia.csv` y extraer el `score_crediticio` de cada segmento en tres Series/arrays separados
2. Calcular estadísticas descriptivas (media, std, mediana) por segmento con `groupby`
3. Aplicar **ANOVA de un factor** con `scipy.stats.f_oneway` — imprimir F y p-valor
4. Si p < 0.05, aplicar **Tukey HSD** con `pairwise_tukeyhsd` para identificar qué pares difieren
5. Graficar un **boxplot** del score por segmento (añadir la media como punto)
6. Redactar una conclusión de 3–4 oraciones para el directorio

> `from scipy import stats`  
> `from statsmodels.stats.multicomp import pairwise_tukeyhsd`

```python
# ============================================================
# EJERCICIO 3.4 — ANOVA + Tukey
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd

RUTA_CSV = 'C:/Users/matic/Data Science/Exam/cartera_crediticia.csv'
df = pd.read_csv(RUTA_CSV)

# 1. Extraer score por segmento
retail      = df.loc[df['segmento'] == 'Retail',      'score_crediticio'] 
pyme        = df.loc[df['segmento'] == 'PYME',        'score_crediticio']
corporativo = df.loc[df['segmento'] == 'Corporativo', 'score_crediticio']


# 2. Estadísticas descriptivas por segmento
print('-'*100)
print("Estadísticas Descriptivas por Segmento:")
print(df.groupby('segmento')['score_crediticio'].agg(['count','mean','std','median']).round(1))
print('-'*100)

# 3. ANOVA de un factor
F, p_anova = stats.f_oneway(retail, pyme, corporativo)
print('ANOVA de un factor:')
print(f"   Estadístico F: {F:.4f} | p-valor: {p_anova:.4e}")
print('-'*100)

# 4. Tukey HSD (si p < 0.05)
tukey = pairwise_tukeyhsd(df['score_crediticio'], df['segmento'], alpha=0.05)
if p_anova < 0.05:
    print("El p-valor del ANOVA es < 0.05. Existen diferencias significativas.")
    print("Aplicando la prueba Post-Hoc de Tukey HSD:\n")
    tukey = pairwise_tukeyhsd(df['score_crediticio'], df['segmento'], alpha=0.05)
    #tukey = pairwise_tukeyhsd(endog=df['score_crediticio'], groups=df['segmento'], alpha=0.05)
    print(tukey)
else:
    print("El p-valor del ANOVA es >= 0.05. No se detectan diferencias significativas entre los segmentos.")
print("-" * 100)

# 5. Boxplot
fig, ax = plt.subplots(figsize=(9, 5))
sns.boxplot(
    data=df, 
    x='segmento', 
    y='score_crediticio', 
    palette='Set2', #Soy conciente de la advertencia que sale , pero prefiero así
    showmeans=True, 
    meanprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"black", "markersize":7}
)
#Títulos:
ax.set_title('Distribución de Score Crediticio por Segmento', fontsize=14, fontweight='bold')
ax.set_xlabel('Segmento', fontsize=12)
ax.set_ylabel('Score Crediticio', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
