## Ejercicio 1: Procesamiento de Imágenes Médicas (Máscaras y `np.where`) 🩻
**Contexto:** En un proyecto de Machine Learning para detección de anomalías pulmonares, los escáneres CT se representan como matrices de densidades de píxeles (valores entre -1000 y 1000). 
**Tarea:**
1. Usando el generador aleatorio moderno (`default_rng`), crea una matriz `escaneo` de tamaño $256 \times 256$ con valores enteros aleatorios entre -1000 y 1000.
2. Los tejidos anómalos suelen tener densidades superiores a 400. Crea una máscara booleana para identificar estos píxeles.
3. Utiliza `np.where` para crear una nueva matriz `mascara_binaria` donde los píxeles sospechosos sean `1` y el resto `0`.
4. Calcula y muestra el porcentaje exacto de píxeles sospechosos en todo el escáner sin usar bucles `for`.

```python
import numpy as np

rng = np.random.default_rng(2004)
np.set_printoptions(precision=4, suppress=True)
np.__version__
```

```python
#Creamos la matriz escaneo:
escaneo = rng.integers(-1000,1000, size = (256,256))
escaneo
```

```python
mask = escaneo > 400
mask
```

```python
mascara_binaria = np.where(mask , 1 ,0)
mascara_binaria
```

```python
porcentaje = mascara_binaria.sum()/mascara_binaria.size if mascara_binaria.size > 0 else None
porcentaje
```

## Ejercicio 2: Puntos de Equilibrio en Sistemas Dinámicos (Álgebra Lineal) 🌀
**Contexto:** Al estudiar sistemas dinámicos lineales del tipo $Ax = b$, necesitas encontrar rápidamente el punto de equilibrio en un espacio multidimensional.
**Tarea:**
1. Define la matriz de coeficientes $A$ de $4 \times 4$ y el vector constante $b$ de tamaño $4$ (inventa valores no triviales usando `np.array`).
2. Utiliza `np.linalg.solve` para encontrar el vector de estado $x$.
3. Demuestra mediante código que tu solución es correcta realizando la multiplicación matricial ($A \times x$) y usando la función `np.allclose` para compararlo con $b$.

```python
#Definimos la matriz
A = np.array([[1,2,3,4],[1,2,10,1],[2,-2,3,10],[0,0,2,-3]])
b = np.array([3,4,0,-1])
A,b

A_random = rng.random((4,4))
b_random = rng.random(4) 
```

```python
#Resolvemos:
try:
    x = np.linalg.solve(A, b)
    print("Solución única encontrada:")
    print(x)
except np.linalg.LinAlgError:
    print("Error: La matriz es singular. No existe solución única.")

try:
    x_random = np.linalg.solve(A_random,b_random)
    print("Solución única encontrada:")
    print(x_random)
except np.linalg.LinAlgError:
    print("Error: La matriz es singular. No existe solución única.")
```

```python
#Comprobamos:
print('Los valores están muy cercanos' if np.allclose(A@x,b,rtol=0.0001) else 'Los valores están muy lejanos')
print('Los valores están muy cercanos' if np.allclose(A_random@x_random,b_random,rtol=0.0001) else 'Los valores están muy lejanos')
```

## Ejercicio 4: Simulación Monte Carlo de Distribución Estelar (Random) 🌌
**Contexto:** Estás simulando la distribución de estrellas en un cúmulo globular esférico.
**Tarea:**
1. Genera una matriz de $100,000 \times 3$ que represente las coordenadas $(x, y, z)$ de posibles estrellas, usando una distribución uniforme entre -5 y 5.
2. El núcleo del cúmulo tiene un radio $R = 2.5$. Usando operaciones vectorizadas, calcula la distancia al cuadrado de cada estrella desde el origen $x^2 + y^2 + z^2$.
3. Crea un filtro para aislar únicamente las estrellas que caen dentro del núcleo esférico.
4. ¿Cuántas estrellas componen el núcleo en esta simulación? Resuélvelo usando la máscara creada.

```python
N=100000
cordenadas = rng.uniform(-5,5, size=(N,3))
cordenadas
```

```python
distancia = cordenadas[:,0]**2 + cordenadas[:,1]**2 + cordenadas[:,2]**2
distancia
```

**Tarea:**
1. Utiliza `np.linspace` para generar 3 arreglos diferentes de 1200 valores cada uno (representando 3 cuerdas distintas).
2. Usa `np.stack` para combinar estas 3 señales en una sola matriz de forma `(3, 1200)` (los canales son las filas).
3. Transpón la matriz resultante para que su forma sea `(1200, 3)` (formato estándar de tiempo $\times$ canales).

## Ejercicio 6: Distancias en Juegos de Mundo Abierto (Broadcasting Avanzado) 🏕️
**Contexto:** En un motor de simulación de supervivencia, necesitas calcular la distancia desde tu base a múltiples recursos (hierro, carbono, etc.) simultáneamente.
**Tarea:**
1. Crea un arreglo `base` con tus coordenadas $(X, Y, Z)$ de forma `(3,)`.
2. Crea una matriz `recursos` con las coordenadas de 50 ítems de forma `(50, 3)`.
3. Sin usar bucles `for`, resta las coordenadas de la base a todos los recursos simultáneamente.
4. Eleva al cuadrado el resultado, suma a través del eje correcto (`axis`) y saca la raíz cuadrada (`np.sqrt`) para obtener un arreglo unidimensional de 50 distancias exactas.

```python
base = rng.uniform(0,1000,size=(3,))
base
```

```python
recursos = rng.uniform(0,1000,size=(50,3))
recursos
```

```python
R = recursos - base.T
R
```

```python
final = np.sqrt((R**2).sum(axis = 1))
final
```

## Ejercicio 7: Mapas de Calor Tácticos (Outliers, `argmax` y `clip`) ⚽
**Contexto:** Analizas el mapa de calor (posesión de balón) en una cuadrícula del campo de fútbol.
**Tarea:**
1. Crea una cuadrícula `campo` de $10 \times 10$ con valores aleatorios extraídos de una distribución normal estándar (`rng.standard_normal`).
2. Algunos valores pueden ser ruido estadístico extremo. Usa `np.clip` para limitar todos los valores del campo a un rango entre -1.5 y 1.5.
3. Encuentra el valor de intensidad máxima de posesión en la cuadrícula ya recortada.
4. Utiliza `np.argmax` y `np.unravel_index` para imprimir la coordenada exacta `(fila, columna)` donde ocurre esta intensidad máxima.

```python
campo = rng.standard_normal(size=(10,10))
campo
```

```python
campo_f = np.clip(campo , -1.5 , 1.5)
campo_f
```

```python
pos = np.unravel_index(campo_f.argmax(), campo_f.shape)
pos
```

```python
print("max global:", campo_f[pos], "en", pos)
```
