## Ejercicio 1: Funciones básicas, parámetros y `return`
**Contexto:** En ciencia de datos, a menudo necesitamos crear métricas calculadas.
**Tarea:** Crea una función llamada `calcular_imc` que reciba dos parámetros: `peso` (en kg) y `altura` (en metros). La función debe calcular el Índice de Masa Corporal (IMC) usando la fórmula: `peso / (altura ** 2)` y retornar el resultado redondeado a 2 decimales usando la función built-in `round()`.

Prueba tu función con un peso de 75 kg y una altura de 1.75 m.

```python
# Escribe tu código aquí:
def calcular_imc(peso,altura):
    imc = peso/(altura**2) if altura > 0 else None
    return round(imc)

peso=75
altura=1.75
print('Su Índice de Masa Corporal (IMC): ',calcular_imc(peso,altura))
```

## Ejercicio 2: Múltiples retornos y parámetros por omisión
**Contexto:** Necesitamos un resumen rápido estadístico de las ventas diarias.
**Tarea:** Crea una función llamada `resumen_ventas` que reciba una lista de números (ventas) y un parámetro por defecto `moneda="USD"`. 
La función debe calcular y retornar tres valores:
1. El total de las ventas (usando `sum`).
2. La venta máxima (usando `max`).
3. Un texto que diga "Moneda: USD" (o la moneda que se pase).

```python
ventas_del_dia = [120, 340, 50, 890, 210]
# Escribe tu código aquí:
def resumen_ventas(ventas,moneda = "USD"):
    total = sum(ventas)
    maxima = max(ventas)
    print("Moneda: ",moneda)
    return total , maxima

resumen_ventas(ventas_del_dia,moneda = "USD")
```

## Ejercicio 3: Limpieza de texto y métodos de string
**Contexto:** Has recibido una base de datos con nombres de clientes mal formateados.
**Tarea:** Escribe una función `limpiar_nombres` que reciba una lista de strings. Debes usar un bucle (o comprensión de listas) y los métodos `.strip()` y `.title()` para devolver una nueva lista con los nombres sin espacios extra a los lados y con la primera letra en mayúscula.

```python
clientes_raw = ["  mArIa GOmez ", "juan perez", "   carlos SANCHEZ "]
# Escribe tu código aquí:
def limpiar_nombres(lista):
    return [s.strip().title() for s in lista]

limpiar_nombres(clientes_raw)
```

## Ejercicio 4: Uso de `zip()`, `filter()` y Funciones `lambda`
**Contexto:** Tienes dos listas, una con nombres de productos y otra con sus respectivos precios. Quieres encontrar solo los productos premium (precio mayor a 100).
**Tarea:** 1. Usa `zip()` para unir los productos y los precios en una lista de tuplas.
2. Usa `filter()` junto con una función `lambda` para filtrar solo aquellas tuplas donde el precio sea mayor a 100.
3. Convierte el resultado a una lista e imprímela.

```python
productos = ["Teclado", "Monitor", "Mouse", "Silla Ergonomica"]
precios = [45, 250, 25, 180]
# Escribe tu código aquí:
def filtro(productos,precios):
    return list(filter(lambda s: s > 100, zip(productos,precios)))

filtro(productos,precios)
    
```

## Ejercicio 5: Comprensión de Listas (List Comprehension)
**Contexto:** Tienes una lista de temperaturas en grados Celsius y necesitas convertirlas a Fahrenheit.
**Tarea:** Utiliza comprensión de listas para crear una nueva lista llamada `temp_fahrenheit`. La fórmula de conversión es: `(C * 9/5) + 32`. Solo debes convertir y guardar las temperaturas que sean mayores a 0 grados Celsius.

```python
temp_celsius = [-5, 0, 12, 25, -10, 30, -100]
# Escribe tu código aquí:
def cel_to_fah(lista):
    return [(s * 9/5) + 32 for s in lista if (s * 9/5) + 32 > 0]

cel_to_fah(temp_celsius)
```

## Ejercicio 6: Manejo de Excepciones (`try`, `except`)
**Contexto:** Al procesar grandes volúmenes de datos, a veces vienen valores nulos o textos donde deberían ir números, lo que rompe los cálculos.
**Tarea:** Crea una función `conversion_segura_entero` que reciba un valor.
- Dentro de un bloque `try`, intenta convertir ese valor a `int()` y retornarlo.
- Usa `except (ValueError, TypeError):` para atrapar errores y retornar `-1` si el valor no se puede convertir.
Prueba tu función con la lista provista usando `map()`.

```python
datos_sucios = ["100", "25", "Error", None, "42"]
# Escribe tu código aquí:
def conversion_segura_entero(valor):
    try:
        return int(valor)   
    except (ValueError,TypeError): 
        return -1

list(map(conversion_segura_entero,datos_sucios))
```

## Ejercicio 1: Calculadora dinámica de precios (`*args` y `**kwargs`) 🛒
* **Contexto:** En retail, los productos pueden tener múltiples descuentos secuenciales y recargos que varían según la temporada[cite: 40].
* [cite_start]**Tarea:** Crea una función llamada `aplicar_ajustes` que reciba un `precio_base`, una cantidad arbitraria de descuentos (`*args` - por ejemplo, 0.10 para un 10% de descuento), y configuraciones adicionales como impuestos o costo de envío (`**kwargs`)[cite: 41].
* **Requisito:** * La función debe primero restar todos los descuentos al precio base, uno tras otro[cite: 42].
  * Luego, si existe la clave impuesto en los `**kwargs`, debe aplicarlo sobre el subtotal[cite: 43].
  * Finalmente, si existe la clave envio, debe sumar ese costo fijo al final[cite: 44].
  * [cite_start]La función debe retornar el precio final[cite: 45].

```python
def aplicar_ajustes(precio_base,*args,**kwargs):
    costo_envio=0
    for i in args:
        precio_base = precio_base*(1-i)
    subtotal=precio_base
    for clave in kwargs:
        if clave == impuesto:
            subtotal=subtotal*(1-kwargs[clave])
        elif clave == envio:
            costo_envio=costo_envio+kwargs[clave]
    precio_final=subtotal + costo_envio
    return precio_final
```

```python
#Definiremos pc como 1%
pc=1/100
aplicar_ajustes(
    1000,
    10*pc,
    5*pc,
    3*pc,
    20*pc,
    impuesto= 18*pc,
    envio= 100
)
```

### COMENTARIOS:

## Ejercicio 2: Limpieza profunda de texto y Comprensión de Listas 🧹
* [cite_start]**Contexto:** Extraes datos de un sistema antiguo y los nombres de las ciudades vienen con espacios extra, mayúsculas incorrectas o incluso campos vacíos[cite: 46].
* **Tarea:** Tienes la lista: `ciudades_sucias = [" lImA", "bogotA ", " BUENos AireS ", "santiago", " "]`[cite: 47].
* [cite_start]**Requisito:** * Usa solo una comprensión de listas para construir una nueva lista de ciudades válidas[cite: 48].
  * [cite_start]Debes eliminar espacios laterales, poner cada ciudad en formato "Título" (la primera letra de cada palabra en mayúscula) y omitir aquellos elementos que, tras limpiarlos, queden como textos vacíos[cite: 49].


```python
ciudades_sucias = [" lImA", "bogotA ", " BUENos AireS ", "santiago", " "]
#SOLUCIÓN:
list(filter(lambda s: s != '',[s.strip().title() for s in ciudades_sucias]))
```

## Ejercicio 3: Transformación en bloque (`map`, `filter`, `lambda`) ⚙️
* **Contexto:** Tienes un flujo de datos de temperaturas de una maquinaria en grados Fahrenheit y necesitas aislar solo las métricas de sobrecalentamiento[cite: 50].
* [cite_start]**Tarea:** Tienes la lista: `temps_f = [85, 110, 32, 104, 50, 120]`[cite: 51].
* **Requisito:** Crea una sola línea de código (asignada a una variable) que encadene funciones integradas para:
  1. Filtrar las temperaturas que sean mayores a 100 grados F (usando `filter` y `lambda`)[cite: 52].
  2. Convertir esas temperaturas filtradas a grados Celsius (usando `map` y `lambda`)[cite: 53].
  3. La fórmula es: `(F - 32) * 5/9`[cite: 54].
  4. [cite_start]Convertir el resultado final a una lista[cite: 55].

```python
temps_f = [85, 110, 32, 104, 50, 120]
#SOLUCIÓN:
list(map(lambda t: (t-32)*5/9 ,filter(lambda t : t > 100,temps_f)))
```

### Ejercicio 4: Pipeline de conversión robusta (`try-except`) 🛡️
* [cite_start]**Contexto:** Un sensor de IoT envía datos numéricos, pero a veces la red falla y envía texto basura o valores nulos que podrían romper tus cálculos[cite: 56].
* **Tarea:** Crea una función `procesar_lecturas` que reciba una lista con datos mixtos, como: `lecturas = ["15.5", "error", None, "22.1", "-5", "sin_conexion"]`[cite: 57].
* [cite_start]**Requisito:** * Itera sobre la lista e intenta convertir cada valor a `float`[cite: 58].
  * [cite_start]Usa un bloque `try-except` para atrapar errores de tipo `ValueError` y `TypeError`[cite: 59].
  * Si la conversión es exitosa, agrega el número a una lista de valores válidos[cite: 60].
  * [cite_start]Si falla, guarda el valor original problemático en una lista separada de anomalías[cite: 61].
  * [cite_start]Retorna ambas listas (por ejemplo, empaquetadas en una tupla)[cite: 62].


```python
lecturas = ["15.5", "error", None, "22.1", "-5", "sin_conexion"]
def procesar_lecturas(lecturas):
    lista_exitosa=[]
    lista_defectuosa=[]
    for lectura in lecturas:
        try: 
            lista_exitosa.append(float(lectura))
        except (ValueError, TypeError):
            lista_defectuosa.append(lectura)
    return print("La lista exitosa es :",lista_exitosa,"La lista defectuosa es :",lista_defectuosa)

procesar_lecturas(lecturas)
```
