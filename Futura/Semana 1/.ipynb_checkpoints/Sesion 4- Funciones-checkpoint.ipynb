## 1) ¿Qué es una función?

Una función es un conjunto de expresiones y sentencias (algoritmo) que se ejecuta cuando es invocada.

Estructura mínima:
```python
def mi_funcion():
    pass
```


Ejemplo base: una función que no recibe parámetros.


````python
def saludar():
    print("Hola, ¿cómo están?")

saludar()
````

## 2) Invocar (llamar) una función

Puntos clave:
- Definir una función no la ejecuta.
- Para ejecutarla, hay que usar `()`.


````python
def funcion_1():
    print("Hola Mundo")

print("Tipo:", type(funcion_1))
funcion_1()
````

Aplicación simple: repetir una acción llamando varias veces.


````python
def ping():
    print("ok")

for _ in range(3):
    ping()
````

## 3) `return`: devolver valores

Cuando una función haga un retorno de datos, éstos pueden ser asignados a una variable.

Punto clave:
- `print()` muestra.
- `return` devuelve.


````python
def funcion_2():
    return "Hola Mundo"

frase = funcion_2()
print(frase)
````

Aplicación: calcular un valor y reutilizarlo en otra operación.


````python
def doble(x):
    return x * 2

resultado = doble(21)
print("Doble:", resultado)
print("Doble + 10:", resultado + 10)
````

## 4) Parámetros

Una función puede recibir datos (parámetros) y trabajar con ellos.


````python
def multiplica_por_5(numero):
    print(f'{numero} x 5 = {numero * 5}')

multiplica_por_5(7)
multiplica_por_5(113)
````

Aplicación (Logística): costo por peso.


````python
def costo_envio(peso_kg, tarifa_por_kg):
    return peso_kg * tarifa_por_kg

print(costo_envio(10, 3.5))
print(costo_envio(2.4, 4.2))
````

## 5) Varios return en una misma función

`return` hace que termine la ejecución de la función cuando aparece.
Esto se usa mucho cuando hay reglas.


````python
def es_par(numero):
    if numero % 2 == 0:
        return 'Número par'
    else:
        return 'Número impar'

print(es_par(2))
print(es_par(3))
````

Aplicación (Operaciones/IoT): alerta por umbral.


````python
def nivel_alerta(valor, umbral_alto=0.9, umbral_medio=0.7):
    if valor >= umbral_alto:
        return "ALTA"
    if valor >= umbral_medio:
        return "MEDIA"
    return "BAJA"

for v in [0.3, 0.75, 0.95]:
    print(v, "->", nivel_alerta(v))
````

## 6) Devolver más de un valor

En Python puedes devolver una tupla y luego desempaquetar.


````python
def cuadrado_y_cubo(numero):
    return numero ** 2, numero ** 3, numero

print(cuadrado_y_cubo(10))
cuad, cubo, num = cuadrado_y_cubo(5)
print(cuad, cubo, num)
````

Aplicación (Data): devolver métricas de una lista.


````python
def metricas_basicas(valores):
    total = sum(valores)
    minimo = min(valores)
    maximo = max(valores)
    promedio = total / len(valores) if len(valores) > 0 else None
    return total, minimo, maximo, promedio

ventas = [120, 80, 200, 40, 150]
t, mi, ma, pr = metricas_basicas(ventas)
print("total/min/max/prom:", t, mi, ma, pr)
````

## 7) Parámetros por omisión (valores por defecto)

Se puede asignar parámetros por defecto.

Regla importante:
- Los parámetros opcionales deben ir después de los obligatorios.


````python
def saludo(nombre, mensaje='Hola'):
    print(mensaje, nombre)

saludo('Luis Garayar')
saludo('Luis Garayar', 'Buenas noches')
saludo(nombre='Luis Garayar')
````

Ejemplo incorrecto (no ejecutar):
```python
def saludo(nombre='Luis Garayar', mensaje):
    print(mensaje, nombre)
```


Aplicación (Retail): precio final con impuesto por defecto.


````python
def precio_final(precio, impuesto=0.18):
    return precio * (1 + impuesto)

print(precio_final(100))
print(precio_final(100, impuesto=0.10))
````

## 15) `map()`
Ejecuta una función sobre cada elemento.


````python
def dup(n):
    return n * 2

list(map(dup, [1, 2, 3, 4]))
````

````python
def mult(a, b):
    return a * b

list(map(mult, [1, 2, 3, 4], [5, 6, 7, 8]))
````

## 16) Parámetros arbitrarios (`*args`, `**kwargs`)
Se usan cuando no sabemos cuántos parámetros llegarán.


````python
def recorrer_parametros_arbitrarios(parametro_fijo, *arbitrarios):
    print(parametro_fijo, " ---fijo")
    for argumento in arbitrarios:
        print(argumento, " ---arbitrario")

recorrer_parametros_arbitrarios('Fixed', 'arbitrario 1', 'arbitrario 2', 'arbitrario 3')
recorrer_parametros_arbitrarios('Fixed')
````

Aplicación (Ventas): sumar n montos.


````python
def suma_arbitraria(a=0, *args):
    total = a
    for i in args:
        total += i
    return total

print(suma_arbitraria(100, 50, 25, 10))
````

Ejemplo con `**kwargs`.


````python
def recorrer_parametros_arbitrarios2(parametro_fijo, *arbitrarios, **kwords):
    print(parametro_fijo, " ---fijo")
    for argumento in arbitrarios:
        print(argumento, " ---arbitrario")
    for clave in kwords:
        print("El valor de", clave, "es", kwords[clave], " ---kword")

recorrer_parametros_arbitrarios2(
    "Fixed",
    "arbitrario 1",
    "arbitrario 2",
    clave1="valor uno",
    clave2="valor dos"
)
````

Aplicación (Configuración): kwargs para parámetros opcionales.


````python
def configurar_modelo(**kwargs):
    defaults = {"lr": 0.1, "max_depth": 3, "seed": 42}
    defaults.update(kwargs)
    return defaults

print(configurar_modelo())
print(configurar_modelo(lr=0.05, max_depth=6))
````

## 17) Desempaquetado de parámetros (`*` y `**`)
Permite pasar listas/tuplas/dicts como argumentos.


````python
def calcular(importe, descuento):
    return importe - (importe * descuento / 100)

print(calcular(2000, 20))

datos = [1500, 10]
print(calcular(*datos))

datos = {"descuento": 10, "importe": 1500}
print(calcular(**datos))
````

Aplicación (Pricing): procesar varios productos con desempaquetado.


````python
productos = [(100, 5), (250, 10), (80, 0)]
for p in productos:
    print(calcular(*p))
````

## 18) Funciones lambda (con ejemplos aplicados)
Lambda: función de una sola expresión.


````python
potencia = lambda x, y: x**y
print(potencia(19, 3))
print(potencia(2, 5))
````

Aplicación (Data): transformar lista con `map` + `lambda`.


````python
nums = [1, 2, 3, 25, 33]
print(list(map(lambda x: x**2, nums)))
````

Aplicación (Texto): normalizar y revertir.


````python
revertir = lambda cadena: cadena[::-1].lower().split()
print(revertir("Luis Garayar"))
````

Aplicación (Calidad de datos): validación rápida.


````python
es_valido = lambda x: (x is not None) and (x >= 0)
datos = [10, None, -3, 5]
print([es_valido(x) for x in datos])
````

Aplicación (Condicional ternario): etiqueta par/impar.


````python
impar2 = lambda numero: 'Es impar' if numero % 2 != 0 else 'Es par'
print(impar2(5))
print(impar2(10))
````

## 20) Manejo de Excepciones en funciones
Para manejo de excepciones se usa `try-except-finally`.


Ejemplos que generan error (dejar comentado y activar si se quiere mostrar):


````python
# 1/0
# 'a'/5
````

````python
def dividir(x, y):
    try:
        resultado = x / y
    except ZeroDivisionError:
        print("¡división por cero!")
    except TypeError:
        print("¡numerador o denominador no son numeros!")
    else:
        print("el resultado es " + str(resultado))
    finally:
        print("Proceso terminado...")

dividir(10, 5)
dividir(10, 0)
dividir("a", 5)
dividir(5, 'b')
````

Aplicación (Datos): convertir a float sin romper el programa.


````python
def to_float_safe(x):
    try:
        return float(x)
    except (ValueError, TypeError):
        return None

raw = ['10', '3.5', 'error', '', None, '-2']
clean = [to_float_safe(x) for x in raw]
print('Clean:', clean)
valid = [x for x in clean if x is not None]
print('Valid:', valid)
print('Promedio:', sum(valid)/len(valid) if valid else None)
````

Ejercicio: agregar elemento una sola vez (con ValueError).


````python
elementos = [1, 5, -2]

def agregar_una_vez(lista, el):
    try:
        if el in lista:
            raise ValueError
        else:
            lista.append(el)
            print(lista)
    except ValueError:
        print("Error: Imposible añadir elementos duplicados =>", el)
    finally:
        print("Proceso terminado...")

agregar_una_vez(elementos, 10)
agregar_una_vez(elementos, -2)
agregar_una_vez(elementos, "Hola")
agregar_una_vez(elementos, 9999)
print(elementos)
````

La sentencia `raise` permite forzar que ocurra una excepción específica.


````python
try:
    raise NameError('Hola')
except NameError:
    print('Ha sucedido una excepción!')

# raise  # esto fuera de un except vuelve a romper
````

## 21) Llamando a un módulo implementado

Este bloque depende de que exista un archivo `modulo_funciones.py` en el mismo directorio.

Aquí no solo importamos: también usamos funciones del módulo en un flujo completo.


````python
try:
    import modulo_funciones as mf   # forma clásica, llama a todas las funciones
    print('Módulo importado OK')
except Exception as e:
    print('No se pudo importar modulo_funciones:', e)
````

````python
try:
    help(mf)
except:
    pass
````

Ejemplo: usar una función del módulo.


````python
try:
    print(mf.sumar_numeros(30, 15, 6))
except Exception as e:
    print(e)
````

Importar una función específica.
Nota: se evita input para no frenar ejecución automática (el mentor puede cambiarlo a input si desea).


````python
try:
    from modulo_funciones import es_par
    numero = 73
    if es_par(numero):
        print('Número par')
    else:
        print('Número impar')
except Exception as e:
    print(e)
````

Importar con alias.


````python
try:
    from modulo_funciones import sumar_numeros as nueva_suma
    print(nueva_suma(15, 35))
except Exception as e:
    print(e)
````

Importar múltiples funciones.


````python
try:
    from modulo_funciones import sumar_numeros, es_par
    print(sumar_numeros(90, 25))
    print(es_par(73))
except Exception as e:
    print(e)
````

Nota:
- No es recomendable usar `from modulo_funciones import *` porque deja sueltas las funciones.
- Mejor utilizar importación clásica: `import modulo_funciones as mf` y luego `mf.funcion()`.


Aplicación final (usando el módulo):
- tomamos una lista de números
- filtramos pares con la función del módulo
- sumamos los pares con la función del módulo

Si el módulo no existe, este bloque simplemente no corre.


````python
try:
    nums = [1, 2, 3, 4, 10, 11, 20]
    pares = [n for n in nums if mf.es_par(n)]
    print('Pares:', pares)
    print('Suma pares:', mf.sumar_numeros(*pares))
except Exception as e:
    print(e)
````
