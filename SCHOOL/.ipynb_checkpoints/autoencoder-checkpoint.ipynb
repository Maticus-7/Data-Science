# Autoencoder


Let us consider the one-dimensonal linear wave equation
\begin{align}
    \partial_t^2 u(x,t) &= c^2 \partial_x^2 u(x,t),
\end{align}
where $c$ is the transport velocity, on a periodic domain $(0, L]\ni x$.

By introducing another variable $v$ we can transform this second order PDE into a first order PDE system.
\begin{align}
    \begin{aligned}
    \partial_t u + \partial_x v &=0,\\
    \partial_t v + c^2 \partial_x u &= 0.
    \end{aligned}
    \tag{2}
\end{align}

💻 <font color='blue'>**Task A1:** </font>
- Show that $u$ solves $(1)$, if it is a part of the solution pair $(u,v)$ of $(2)$.
- For a given solution $u$ of $(1)$, construct a function $v$ such that the pair $(u,v)$ is a solution of $(2)$.
- Proof that 
\begin{align*}
    \begin{pmatrix}
    u(t, x)\\
    v(t,x)
    \end{pmatrix}=
    \begin{pmatrix}
    q_+(x-ct) + q_-(x+ct)\\
    c (q_+(x-ct) -  q_-(x+ct))
    \end{pmatrix}
\end{align*}
   solves $(2)$ for any two functions $q_+,q_- \in C^2(\mathbb{R})$.

We import all necessary packages.

```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
```

## Transport quantities

We start with an example of only one transported quantity, i.e, $q_{-}=0$. The quantity of choice is a sharp gaussian pulse $q_+(x)=\exp(-(x-x_0)^2/\delta^2)$.

```python
L = 1 # domain length
c = 1 # wave speed
x0 = 0.1   
delta = L / 50  # width of the Gaussian
def q_plus(x):
    return np.exp(-(x-x0)**2/delta**2)

def q_minus(x):
    return np.zeros_like(x)
```

Define analytical solution

```python
def u(t,x):
    return q_plus((x-t) % L) + q_minus((x+t) % L)

def v(t,x):
    return q_plus((x-t) % L) - q_minus((x+t) % L)
```

Build snapshot matrix

```python
nx = 200
nt = 250
tspan = np.linspace(0.0, 1.25*L/c, nt)
xspan = np.linspace(0, L, nx)
t, x = np.meshgrid(tspan, xspan)
U = u(t, x)
V = v(t, x) 
X = np.stack((U, V), axis=-1)  # (nx, nt, 2)

fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharex=True, sharey=True)
im1 = axes[0].imshow(
    X[...,0],
    extent=[tspan[0], tspan[-1], xspan[0], xspan[-1]],
    origin='lower',
    aspect='auto'
)
axes[0].set_title('Density $u(t,x)$')
axes[0].set_xlabel("Time $t$")
axes[0].set_ylabel("Position $x$")

im2 = axes[1].imshow(
    X[...,1],   
    extent=[tspan[0], tspan[-1], xspan[0], xspan[-1]],
    origin='lower',
    aspect='auto'
)
axes[1].set_title('Velocity $v(t,x)$')
axes[1].set_xlabel("Time $t$")
axes[1].set_ylabel("Position $x$")

fig.colorbar(im1, ax=axes[0], label="$u(t,x)$")
fig.colorbar(im2, ax=axes[1], label="$v(t,x)$")
plt.tight_layout()
plt.show()
```

We take a look at the singular values of $\mathbf{X}_u$ and $\mathbf{X}_v$.

```python
U_u, S_u, V_uT = np.linalg.svd(X[...,0], full_matrices=False)
plt.semilogy(S_u / S_u[0], 'o')
plt.title("Singular values of $X_u$")
plt.xlabel("Index $i$")
plt.ylabel(r"Singular value $\sigma_i$ (log scale)")
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```

Since the reconstruction is poor for low ranks, we introduce a nonlinear model order reduction method in autoencoders.
## Autoencoders
An autoencoder consists of two parts. An encoder $\mathcal{E}:\mathbb{R}^n \to \mathbb{R}^r$ that maps the data to a latent space and a decoder $\mathcal{D}: \mathbb{R}^r \to \mathbb{R}^n$ that maps it back to the original space. The autoencoder $\mathbf{\Psi}:\mathbb{R}^n \to \mathbb{R}^n$ is the composition of encoder and decoder $\mathbf{\Psi} = \mathcal{D} \circ \mathcal{E}$.

The autoencoder loss is given by: 
\begin{align*}
    \mathcal{L} = \Vert \mathcal{D}(\mathcal{E}(\mathbf{X}))-\mathbf{X}\Vert
\end{align*}

💻 <font color='blue'>**Task A3:** Implement a simple autoencoder architecture.</font>
- It should have the following inputs
    - input dimension
    - latent dimension
    - an array that contains the width of all hidden layers
    - the choice of activation function as a string, i.e., `'relu'`
- It should have a method that encodes the data.
- It should have a method that decodes the latent data. The architecture of the decoder should be the reverse architecture of the encoder.
- It should have a forward pass that first encodes and then decodes its input.
