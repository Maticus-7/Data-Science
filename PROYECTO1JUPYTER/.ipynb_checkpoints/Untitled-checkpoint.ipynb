# Puntos mínimos globales y locales

<div style="background: #f8f9fa; padding: 15px; border-left: 5px solid #2e6f9a; margin: 10px 0; color: #333;">
<strong style="color: #2e6f9a;">Definición(Punto mínimo Global):</strong><br>
Un punto $x^* \in S$ se le dice Punto mínimo Global de $f$ en S si: 
    $$f(x*)\leq f(x), \forall x \in S$$
Y el valor $f(x*)$ es dicho como minimo global de f en S.
El punto $x^{**}\in S$ se le dice Punto mínimo Global estricto de $f$ en S si:
    $$f(x^{**})< f(x), \forall x \in S,x\neq x^{**}$$
</div>

```python
from scipy.optimize import minimize

# 1. Define la función objetivo que quieres minimizar
def objetivo(x):
    # Ejemplo: Una función cuadrática simple (mínimo en (0,0,0))
    # pero la restricción esférica evitará llegar a ese punto.
    return x[0]**2 + 2*x[1]**2 + 3*x[2]**2

# 2. Define la restricción esférica: ||x||^2 - R^2 = 0
radio = 1.0
def restriccion_esfera(x):
    return x[0]**2 + x[1]**2 + x[2]**2 - radio**2

# 3. Configura el diccionario de restricción para 'scipy.optimize'
constraints = ({'type': 'eq', 'fun': restriccion_esfera})

# 4. Proporciona un punto inicial (debe estar en la esfera)
#    Por ejemplo: x0 = [1, 0, 0] es un punto válido en la esfera unitaria.
x0 = np.array([1.0, 0.0, 0.0])

# 5. Ejecuta la optimización
#    El método 'SLSQP' es una buena opción para problemas con restricciones.
resultado = minimize(objetivo, x0, method='SLSQP', constraints=constraints)

# 6. Muestra los resultados
print("Resultado de la optimización:")
print(f"  ¿Éxito? {resultado.success}")
print(f"  Punto óptimo (en la esfera): {resultado.x}")
print(f"  Norma del punto óptimo: {np.linalg.norm(resultado.x):.6f}")  # Debe ser ~1.0
print(f"  Valor mínimo de la función: {resultado.fun:.6f}")
```

<div style="background: #f8f9fa; padding: 15px; border-left: 5px solid #2e6f9a; margin: 10px 0; color: #333;">
<strong style="color: #2e6f9a;">Definición(Punto mínimo Local):</strong><br>
Un punto $x^* \in S$ se le dice Punto mínimo Global de $f$ en S si: 
    $$f(x*)\leq f(x), \forall x \in S$$
Y el valor $f(x*)$ es dicho como minimo global de f en S.
El punto $x^{**}\in S$ se le dice Punto mínimo Global estricto de $f$ en S si:
    $$f(x^{**})< f(x), \forall x \in S,x\neq x^{**}$$
</div>

<div style="background: #f8f9fa; padding: 15px; border-left: 5px solid #2e6f9a; margin: 10px 0; color: #333;">
<strong style="color: #2e6f9a;">Teorema (Valor Medio):</strong><br>
Si $f$ es continua en $[a,b]$ y diferenciable en $(a,b)$, entonces existe $c \in (a,b)$ tal que:
$$f'(c) = \frac{f(b)-f(a)}{b-a}$$
</div>

```python
# Ejemplo:
import numpy as np
from scipy.optimize import minimize
from IPython.display import display, Math, Latex

# Para la función: f(x)=(x_1)^2-(x_2)
#Definimos la función
def funcion(x):
    return x[0]**2-x[1]
#Definimos la restricción
def restriccion(x):
    return -x[1]
#Configuramos el diccionario de restricción
constraints = ({'type': 'ineq','fun' : restriccion})
#Proponemos un punto inicial x0
x0= np.array([1,0])
#Ejecutamos
result=minimize(funcion,x0,method='SLSQP',constraints=constraints)
#Mostramos
print('Encontremos el mínimo de la función:')
display(Math(r'f(\mathbf{x,y}) = x_1^2-x_2'))
print('Con la restrición:')
display(Math(r'S = \{ x \in \mathbb{R}^2 /x_2 \leq 0 \}'))
print(f"  Punto óptimo : {result.x}")
print(f"  Valor mínimo de la función: {result.fun:.6f}")
```
