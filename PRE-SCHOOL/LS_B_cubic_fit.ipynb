<font color='blue'> **For this task, we import existing data and try to fit them with different models.**</font>

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('LS_data_file.npy')

x_data = data[0]
y_data = data[1]
```

<font color='blue'> **Lets first try fitting a linear equation to our data.**</font>
We want to find a function $f(x)=c_1 \; x + c_2$ that approximates our data $(x, y_{linear})$. Now you can use directly the built-in Python tool (https://numpy.org/doc/stable/reference/generated/numpy.linalg.lstsq.html).

```python
A_linear = np.vstack([x_data, np.ones(len(x_data))]).T

coef_linear, *_ = np.linalg.lstsq(A_linear, y_data , rcond=None)

y_LS_linear =  A_linear @ coef_linear 
```

<font color='blue'> **Plot your result:**</font>

```python
plt.figure()
plt.scatter(x_data, y_data, label='Data')
plt.plot( x_data, y_LS_linear , color='red', label=' Linear Fit')
plt.xlabel("x"); plt.ylabel("y"); plt.grid(True)
plt.legend()
plt.show()
```

<font color='blue'> **Since the linear fit is not really good, we will add more terms.**</font>
We will try to fit a cubic function to the data, $f(x)=c_1 \; x^3 +c_2 \; x^2 + c_3 \; x  + c_4$.

```python
A_cubic = np.vstack( [x_data,x_data**2,x_data**3,np.ones(len(x_data))]).T

coef_cubic, *_ = np.linalg.lstsq(A_cubic, y_data, rcond=None)

y_LS_cubic =  A_cubic @ coef_cubic
```

<font color='blue'> **Plot your new result:**</font>

```python
plt.figure()
plt.scatter(x_data, y_data, label='Data')
plt.plot( x_data, y_LS_cubic , color='red', label=' Cubic Fit')
plt.xlabel("x"); plt.ylabel("y"); plt.grid(True)
plt.legend()
plt.show()
```

<font color='blue'> **What do you see? Compare to the result from the linear fit.**</font>

```python
plt.figure()
plt.scatter(x_data, y_data, label='Data')
plt.plot( x_data, y_LS_cubic , color='red', label=' Cubic Fit')
plt.plot( x_data, y_LS_linear , color = 'g', label= 'Linear Fi')
plt.xlabel("x"); plt.ylabel("y"); plt.grid(True)
plt.legend()
plt.show()
```
